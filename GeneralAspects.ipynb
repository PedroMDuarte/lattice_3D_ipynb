## Uncompensated lattice, 47 um waist.  Analytic profiles

```python
from matplotlib import rc
rc('font',**{'family':'serif'})
rc('text', usetex=True)

fig = plt.figure(figsize=(5.5,2.9))
gs = matplotlib.gridspec.GridSpec( 1,1, \
        wspace=0.25, left=0.12, right=0.55, bottom=0.18, top=0.9)
ax = fig.add_subplot( gs[0,0])

def vL( s0, wL, r ):
    return s0*np.exp( - 4*r**2 / ( 3. * wL**2 ) ) 

def vLenv( s0, wL, r ):
    return -3.*vL( s0, wL, r) 

def vLenvTop( s0, wL, r):
    return vLenv( s0, wL, r) + vL( s0, wL, r)

def vLatt( s0, wL, r):
    return vLenv( s0, wL, r) + vL( s0, wL, r) *  ( sin(2.*np.pi * r111 / 1.064 /  10. )**2 )

def band_bot( s0, wL, r):
    return vLenv( s0, wL, r) + 3.*np.sqrt( vL( s0, wL, r) ) 

def band_top( s0, wL, r):
    return vLenv( s0, wL, r) + 3.*np.sqrt( vL( s0, wL, r) )  \
           + 48./np.sqrt(np.pi) * ( vL(s0,wL,r)**(3./4.)) * np.exp( -2.*np.sqrt( vL(s0,wL,r)) )
    
def band_1st( s0, wL, r):
    return vLenv( s0, wL, r) + 5.*np.sqrt( vL( s0, wL, r) )  

    
r111 = np.linspace( -80., 80., 540 ) 

s0 = 10.
wL = 47. 

ax.plot( r111, vLenv( s0, wL, r111 ), color='black', lw=2. , \
        label='$V_{L,\mathrm{env}}(r_{\mathrm{d}})$') 
ax.plot( r111, vLenvTop( s0, wL, r111 ), '--',color='black', lw=2., \
        label='$V_{L,\mathrm{env}}(r_{\mathrm{d}}) + s(r_{\mathrm{d}})$' ) 
ax.plot( r111, vLatt( s0, wL, r111 ), color='gray', lw=0.8, alpha=0.5, label=r'$\mathrm{Lattice}\ \lambda\!\times\!\!10$' ) 
ax.plot( r111, band_bot( s0, wL, r111), color='blue', lw=1.4, label=r'$V_{L,\mathrm{env}}(r_{\mathrm{d}}) +\frac{3}{2} \hbar \omega_{0} $')
ax.plot( r111, band_top( s0, wL, r111), color='blue', lw=1.4, label=r'$V_{L,\mathrm{env}}(r_{\mathrm{d}}) +\frac{3}{2} \hbar \omega_{0} + W$')

ax.fill_between( r111, band_bot( s0, wL, r111), band_top( s0, wL, r111), color='blue', alpha=0.5) 

ax.plot( r111, band_1st( s0, wL, r111), color='red', lw=1.4, label=r'$V_{L,\mathrm{env}}(r_{\mathrm{d}}) +\frac{5}{2} \hbar \omega_{0} $')
ax.axhline( band_1st( s0, wL, 0.), color='red', lw=0.8, alpha=0.3)
#ax.axvline( x=24.2, ymin=band_1st( s0, wL, 0.), ymax=-5., color='red', lw=0.8, alpha=0.5) 
ax.axvline( x=22.2,  color='red', lw=0.8, alpha=0.3) 

ax.annotate("",
            xy=(24.2, -5.), xycoords='data',
            xytext=(-1.5, -5.), textcoords='data',
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="arc3",color='red'), color='red'
            )
ax.text( 10., -4., '$r_{1}$', ha='center', va='center', color='red')

ax.grid()


ax.set_ylabel('$E_{R}$')
ax.set_xlabel('$\mu\mathrm{m\ along\ 111}$')

ax.legend( bbox_to_anchor=(1.01,1.00), \
        loc='upper left', numpoints=2, \
        prop={'size':12}, handlelength=2.2, handletextpad=0.5 )

fig.savefig('GeneralAspects_figures/lattice_general.png', dpi=300)
```

## Maximum locked number vs.  lock depth.  For various lattice beam waists 

```python
from matplotlib import rc
rc('font',**{'family':'serif'})
rc('text', usetex=True)

fig = plt.figure(figsize=(3.8,2.9))
gs = matplotlib.gridspec.GridSpec( 1,1, \
        wspace=0.25, left=0.25, right=0.95, bottom=0.15, top=0.95)
ax = fig.add_subplot( gs[0,0])


from scipy import optimize

def band_diff( s0, wL, r111):
    return band_bot( s0, wL, r111) - band_bot( s0, wL, 0.) \
        - ( band_1st( s0, wL, 0. )- band_bot( s0, wL, 0.) )

def find_r1( s0, wL ):
    f = lambda x : band_diff( s0, wL, x ) 
    return optimize.brentq( f , 0., 200., )


def band_diff_comp( s0, wL, g0, wC,r111):
    return band_bot( s0, wL, r111) - band_bot( s0, wL, 0.) \
              + 3.*vL( g0, wC, r111) - 3.*vL( g0, wC, 0.) \
        - ( band_1st( s0, wL, 0. )- band_bot( s0, wL, 0.) )
def find_r1_comp( s0, wL, g0, wC):
    f = lambda x : band_diff_comp( s0, wL, g0, wC, x ) 
    return optimize.brentq( f , 0., 400., )


ss = np.linspace( 10., 60., 20 ) 

wLs = [47., 54., 62.,70.]
for wL in wLs:
    r1 = []
    for s in ss:
        r1.append( find_r1( s,wL))
    r1 = np.array( r1 ) 
    ax.plot(  ss,  4./3. * np.pi * ( r1/ (1.064/2.) )**3, label='$w_{L}=%d\,\mu\mathrm{m}$'%wL, lw=2 )
    
wC = 40. 
wLs = [47.]
for wL in wLs:
    r1 = []
    for s in ss:
        r1.append( find_r1_comp( s,wL, 3.0, wC))
    r1 = np.array( r1 ) 
    ax.plot(  ss,  4./3. * np.pi * ( r1/ (1.064/2.) )**3, ':', label='$w_{L}=%d\,\mu\mathrm{m},\ \mathrm{comp.}$'%wL, lw=2 )
    
    
ax.grid()
ax.set_ylim( 0., 500000)

ax.set_xlabel('$\mathrm{Lattice\ depth\ } (E_{R})$')  
ax.set_ylabel(r'$N_{1} =  \frac{4}{3} \pi \left( \frac{ r_{1} }{ a  }  \right)^{3}$', fontsize=12)

ax.legend( bbox_to_anchor=(1.05,1.00), \
        loc='upper right', numpoints=1, \
        prop={'size':8}, handlelength=1.1, handletextpad=0.5 )


fig.savefig('GeneralAspects_figures/lattice_lock.png', dpi=300)
```

## Tunneling rate vs. Lattice depth

```python
from matplotlib import rc
rc('font',**{'family':'serif'})
rc('text', usetex=True)

fig = plt.figure(figsize=(3.8,2.9))
gs = matplotlib.gridspec.GridSpec( 1,1, \
        wspace=0.25, left=0.22, right=0.95, bottom=0.15, top=0.95)
ax = fig.add_subplot( gs[0,0])

ss = np.linspace( 10., 60., 20 ) 

def tunnel( s ):
    return 4./np.sqrt(np.pi) * ( s**(3./4.)) * np.exp( -2.*np.sqrt( s) ) * 29 


ax.plot( ss , tunnel(ss), lw=2.5 )

ax.grid()
ax.set_yscale('log')
ax.set_ylabel('$\mathrm{Tunneling\ rate\ (kHz)}$')
ax.set_xlabel('$\mathrm{Lattice\ depth\ } (E_{R})$')  

ax.axhline( 0.02/ 10 , color='black')


fig.savefig('GeneralAspects_figures/lattice_tunneling.png', dpi=300)
```

## Characteristic energy vs.  beam waist for 500,000 atoms

```python
from matplotlib import rc
rc('font',**{'family':'serif'})
rc('text', usetex=True)

fig = plt.figure(figsize=(3.2,2.6))
gs = matplotlib.gridspec.GridSpec( 1,1, \
        wspace=0.25, left=0.18, right=0.95, bottom=0.18, top=0.9)
ax = fig.add_subplot( gs[0,0])

def alpha( s0, wL):
    return ( 4*s0 - 2*np.sqrt(s0) ) / ( (wL/(1.064/2.))**2.)  

def t( s0 ):
    return 4./np.sqrt(np.pi) * ( s0**(3./4.)) * np.exp( -2.*np.sqrt( s0) )

def Et( s0, wL, Natoms):
    return alpha(s0,wL) * (  3.*Natoms / (8.*np.pi) )**(2./3.)  / t(s0 ) 


wLs = np.linspace( 60., 200., 100 )
#plt.plot( wLs,  alpha( 10., wLs) ) 
ax.plot( wLs, Et( 10., wLs, 500000 ), lw=3) 

ax.set_xlabel('$w_{L}\ (\mu\mathrm{m})$')
ax.set_ylabel('$E_{\mathrm{t}} / t $')
ax.grid()
ax.axhline( 40., color='black', alpha=0.6)

fig.savefig('GeneralAspects_figures/Et-waist.png', dpi=190)
```
