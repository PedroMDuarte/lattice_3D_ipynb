Example to make a chemical potential cut at fixed U, starting from a filling of 1 particle per site.   Compare with Fig 5. in Fuchs et. al. PRL 106, 030401 (2011)

```python
def trap_fixed_U( U ):
    mu_ = np.linspace( -20, U/2, 100)
    U_  = np.ones_like( mu_ ) * U
    return np.transpose(np.vstack(( U_, mu_)))

for Temp in [1.6, 2.6, 4.6, 10.8 ]:
    fentr = getHTSEInterp2( Temp, name="entropy")
    xy = trap_fixed_U( 8. )
    z = fentr(xy)
    plt.plot( xy[:,1], z ) 
```

Make a plot of the density as a function of chemical potential for various interaction strenghts and temperatures

```python
from matplotlib import rc
rc('font',**{'family':'serif'})
rc('text', usetex=True)

def mu_cut( U ):
    mu_ = np.linspace( -14, 32, 100)
    U_  = np.ones_like( mu_ ) * U
    return np.transpose(np.vstack(( U_, mu_)))

def plotCompare( ax, quantity, Uval=8. ):
    for Temp in [1.8, 3.6, 5.4]:
        fHq = getHTSEInterp2( Temp, name=quantity)
        xy = mu_cut( Uval )
        zH = fHq(xy)
        ax.plot( xy[:,1], zH,lw=2, label = "$T/t=%.1f$"%Temp) 
            
def HTSEshow( qty = 'density', label='atoms per site' ):
        
    fig = plt.figure(figsize=(9.6,2.9))
    gs = matplotlib.gridspec.GridSpec( 1,3, wspace=0.25, left=0.06, right=0.85, bottom=0.18, top=0.9)
    
    ax0 = fig.add_subplot( gs[0,0])
    ax1 = fig.add_subplot( gs[0,1])
    ax2 = fig.add_subplot( gs[0,2])
    
    ax0.set_ylabel(label,fontsize=14,rotation=90,labelpad=5)
   
    axs = [ax0,ax1,ax2]
    for ax in axs:
        ax.set_xlabel("$\mu/t$",fontsize=16)
        ax.set_xlim( -12., 32.)
        #ax.set_ylim( -12., 48.)
        ax.grid()
        
        ax.xaxis.set_minor_locator(matplotlib.ticker.MultipleLocator(2))
        ax.grid(which='minor')
        
    for i,Uval in enumerate([8., 16., 24.]):
        plotCompare(axs[i], qty, Uval=Uval)
        axs[i].set_title('$U/t = %d$'%Uval, fontsize=18, ha='center')
    
    ax2.legend( bbox_to_anchor=(1.03,1.00), \
                loc='upper left', numpoints=1, \
                 prop={'size':12}, handlelength=1.1, handletextpad=0.5 )
    
    #gs.tight_layout(fig, rect=[0.,0.,0.9,0.92])
    fig.savefig('HubbardPhaseDiagram_figures/HTSE_'+qty+'_U.png',dpi=180)
```

Make a plot of entropy per particle as a function of density for various interactions and temperatures

```python
def mu_cut( U ):
    mu_ = np.linspace( -24, 48, 100)
    U_  = np.ones_like( mu_ ) * U
    return np.transpose(np.vstack(( U_, mu_)))

def plotCompare( ax, quantityY, quantityX, Uval=8. ):
    for Temp in [1.8, 3.6, 5.4]:
        fHqY = getHTSEInterp2( Temp, name=quantityY)
        fHqX = getHTSEInterp2( Temp, name=quantityX)
        xy = mu_cut( Uval )
        zHY = fHqY(xy)
        zHX = fHqX(xy) 
        ax.plot( zHX, zHY,lw=2, label = "$T/t=%.1f$"%Temp) 
            
def HTSEshow( ):
        
    fig = plt.figure(figsize=(10.4,3.2))
    gs = matplotlib.gridspec.GridSpec( 1,3, wspace=0.25, left=0.06, right=0.85, bottom=0.18, top=0.9)
    
    ax0 = fig.add_subplot( gs[0,0])
    ax1 = fig.add_subplot( gs[0,1])
    ax2 = fig.add_subplot( gs[0,2])
    
    #ax0.set_ylabel("Atoms per site",fontsize=14,rotation=90,labelpad=5)
    ax0.set_ylabel("entropy per site ($k_{B}$)",fontsize=14,rotation=90,labelpad=18)
   
    axs = [ax0,ax1,ax2]
    for ax in axs:
        #ax.set_xlabel("$\mu/t$",fontsize=16)
        ax.set_xlabel("Atoms per site",fontsize=14)
        #ax.set_xlim( -12., 32.)
        ax.set_ylim( 0., 1.5)
        ax.grid()
        
        #ax.xaxis.set_minor_locator(matplotlib.ticker.MultipleLocator(0.5))
        ax.xaxis.set_major_locator(matplotlib.ticker.MaxNLocator(4))
        ax.yaxis.set_major_locator(matplotlib.ticker.MaxNLocator(6, prune='lower'))
        ax.grid(which='minor')
        ax.tick_params(axis='both', which='major', labelsize=13)
        
    for i,Uval in enumerate([8., 16., 24.]):
        plotCompare(axs[i], "entropy", "density", Uval=Uval)
        axs[i].set_title('$U/t = %d$'%Uval, fontsize=20, ha='center')
    
    ax2.legend( bbox_to_anchor=(1.03,1.00), \
                loc='upper left', numpoints=1, \
                 prop={'size':15}, handlelength=1.1, handletextpad=0.5 )
    
    #gs.tight_layout(fig, rect=[0.,0.,0.88,0.98])
    fig.savefig('HubbardPhaseDiagram_figures/HTSE_EntropyPerSite_U.png',dpi=180)
    
HTSEshow()
```
