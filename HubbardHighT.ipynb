```python
from IPython.display import display

from sympy.interactive import printing
printing.init_printing()

from __future__ import division
import sympy as sym
from sympy import *


import matplotlib
matplotlib.rc('font',**{'family':'serif'})
```

```python
z0, z, u, B, t , U= symbols(r"z_{0} z u \beta t U")
m=6.
grand = -sym.log( z0 ) / B - B * (t/z0)**2 * m * ( z + z**3 * u + 2*z**2 * (1-u) / (B*U))
grand
```

```python
T, mu= symbols(r"T \mu")
B = 1 / T 
z = sym.exp( B*mu )
u = sym.exp(-B*U ) 

z0 = 1 + 2*z + z**2 * u
grand = -sym.log( z0 ) / B - B * (t/z0)**2 * m * ( z + z**3 * u + 2*z**2 * (1-u) / (B*U))
grand
```

```python
grand = grand.subs(t,1)
grand 
```

```python
sym.simplify(grand)
```

```python
Density = -1 * diff( grand, mu )
Density
```

```python
Doublons = diff( grand, U )
Doublons
```

```python
Entropy = -1* diff( grand, T)
Entropy
```

```python
DensFluc = T* diff( -1* diff( grand, mu), mu )
DensFluc
```

```python
import matplotlib.mlab as ml

def phasediagram( Tval ):
    mu_set = np.linspace(-12, 48, 83 )
    U_set  = np.linspace(-12, 48, 83 )
    
    n = []
    n2= []
    d = []
    s = []
    
    
    muIdx = []
    UIdx  = []
    
    for i,muval in enumerate(mu_set):
        for j,Uval in enumerate(U_set):
            dens = Density.subs(( (T, Tval), (mu,muval), (U,Uval) ))
            denf = DensFluc.subs(( (T, Tval), (mu,muval), (U,Uval) ))
            doub = Doublons.subs(( (T, Tval), (mu,muval), (U,Uval) ))
            entr = Entropy.subs(( (T, Tval), (mu,muval), (U,Uval) ))
            if "nan" in [str(dens), str(doub), str(entr)]:
                continue
            else:
                muIdx.append( muval)
                UIdx.append( Uval)
                n.append( dens )
                n2.append( denf )
                d.append( doub )
                s.append( entr )

    muIdx = np.array(muIdx)
    UIdx  = np.array( UIdx) 
    n = np.array(n)
    n2= np.array(n2)
    d = np.array(d)
    s = np.array(s)
    
    ni=500.
    xi = np.linspace(U_set.min(), U_set.max(), ni)
    yi = np.linspace(mu_set.min(), mu_set.max(), ni)

    
    zn = ml.griddata(UIdx, muIdx, n, xi,yi)
    zn2 = ml.griddata(UIdx, muIdx, n2, xi,yi)
    zd = ml.griddata(UIdx, muIdx, d, xi,yi)
    zs = ml.griddata(UIdx, muIdx, s, xi,yi)
    return xi, yi, UIdx, muIdx, zn, zn2, zd, zs
            
```

```python
xi, yi, x, y, zn, zn2, zd, zs = phasediagram(2.5)
```
