# Band structure calculation for a 1D Lattice


We start by defining the function that calculates the truncated Hamiltonian matrix.  I call this Hamiltonian, but it is really the set of liear equations for the coefficients that describe the wave functions in terms of plane waves.  

```python
def H0( Npoints, q, v0 ): 
    def f(i,j):
        if i==j:
            return (q+2*(i-Npoints))**2 + v0/2 
        elif i-j == 1:
            return v0/4
        elif j-i == 1:
            return v0/4 
        else:
            return 0.
    mdim = (2*Npoints+1) 
    matrix = sym.Matrix(mdim,mdim,f)
    return matrix

q, v0 = sym.symbols("q V_{0}")
H0(2,q,v0)
```

With this matrix in hand we can diagonalize it and get the eigenvalues, the eigenvalues of this matrix correspond to the energy for the given quasimomentum, lattice depth, and band index.  At the same time we define the harmonic oscillator states when each lattice site is considered as an independent harmonic oscillator potential. 

```python
def eigenvalsH0( q, v0,  NPoints=5):
    # NPoints is the truncation value of hamiltonian matrix
    q_Sym, v0_Sym = sym.symbols("q V_{0}")
    h0 = np.array( np.array( H0(NPoints,q_Sym,v0_Sym).subs( [(q_Sym,q), (v0_Sym,v0)] )), np.float )
    return np.sort( np.linalg.eigvalsh(h0) )
def bandH0( NBand, qs, v0, NPoints=5):
    Eq = np.array( [ eigenvalsH0( q, v0, NPoints=NPoints)[NBand] for q in qs] )
    return Eq

def hosc( NBand, v0 ):
    return 2* np.sqrt(  v0 ) * (NBand + 0.5)
```

We define a function that allows us to calculate the band width for a certain lattice depth

```python
def bands1d( v0, NBand=0 ):
    Eq0 = eigenvalsH0( 0., v0)[NBand]
    Eq1 = eigenvalsH0( 1., v0)[NBand]
    if Eq0 > Eq1:
        return np.array([Eq1,Eq0])
    else:
        return np.array([Eq0,Eq1])
def bands1dvec( v0, NBand=0):
    band = numpy.empty( (len(v0), 2) ) 
    for i,v in enumerate(v0):
        band[i] = bands1d(v, NBand)
    return band
```

# Calculation of eigenstates for a 1D lattice


To calculate the eigenstates in 1D we diagonlize the Hamiltonian matrix and use the set of eigenvectors to build up the eigentates as a sum of plane waves

## We can build up the eigenstates using arrays:

```python
def eigenfunsH0( q, v0, x, NPoints=5 ):
    # NPoints is the truncation value of hamiltonian matrix
    q_Sym, v0_Sym = sym.symbols("q V_{0}")
    h0 = np.array( np.array( H0(NPoints,q_Sym,v0_Sym).subs( [(q_Sym,q), (v0_Sym,v0)] )), np.float )
    w,v =  numpy.linalg.eigh(h0)
    a = 1.0 # We use x in units of the lattice spacing
    k = np.pi/a 
    
    # m are the indices for the reciprocal lattice vectors
    m = np.linspace(-NPoints,NPoints,2*NPoints+1)
    
    # The sum of plane waves is carried out for each of the bands
    psi_bands = []
    for i in np.argsort(w):
        coefs = v[:,i][:,None]
        #if i==0:
        #    print coefs
        planewaves =  coefs * np.exp( 1j*(2*k)*np.outer(m,x) ) * np.exp(1j*q*k*x)
        psi = planewaves.sum(axis=0)
        
        # Adjust the wavefunctions so they have matching phases at x0
        # where x0 is the position of a lattice site. 
        #
        # This will turn out to be important when defining the Wannier 
        # wavefunctions. 
        #
        x0 = 0.5 
        phase = -1*np.exp( 1j* np.angle( np.sum(coefs * np.exp( 1j*(2*k)*m*x0 ) * np.exp(1j*q*k*x0))))
        
        psi = psi / phase
        psi_bands.append((psi,w[i]))
    return psi_bands   
    
```

## We can also build them up symbollicaly, but this is too slow

```python
x_Sym, qq_Sym, k_Sym = sym.symbols("x q k")
a = 1.0 # We use x in units of the lattice spacing
k = np.pi/a 

def eigenfunsH0_Sym( q, v0, NPoints=5 ):
    # NPoints is the truncation value of hamiltonian matrix
    q_Sym, v0_Sym = sym.symbols("q V_{0}")
    h0 = np.array( np.array( H0(NPoints,q_Sym,v0_Sym).subs( [(q_Sym,q), (v0_Sym,v0)] )), np.float )
    w,v =  numpy.linalg.eigh(h0)
    
    
    # m are the indices for the reciprocal lattice vectors
    m = np.linspace(-NPoints,NPoints,2*NPoints+1)
    
    # The sum of plane waves is carried out for each of the bands
    psi_bands = []
    
    for i in np.argsort(w):
        coefs = v[:,i][:,None]
        
        psi = 0.
        for j,c in enumerate(coefs):
            c = float(c)
            planewave = sym.exp( 1j*(2*k_Sym)*m[j]*x_Sym ) * sym.exp(1j*qq_Sym*k_Sym*x_Sym) 
            #sym.pprint(  planewave, use_unicode=True )
            psi += c * planewave
        
        # Adjust the wavefunctions so they have matching phases at x0
        # where x0 is the position of a lattice site. 
        #
        # This will turn out to be important when defining the Wannier 
        # wavefunctions. 
        #
        x0 = 0.5 
        phase = np.exp( 1j* np.angle( complex(sym.N( psi, subs = { x_Sym:x0, k_Sym:k, qq_Sym:q } ) ) ))
               
        psi = psi / phase
        psi = psi.subs( k_Sym, np.pi ).subs( qq_Sym, 0. )
        #if i == 0:
        #    sym.pprint(psi)
        psi_bands.append(( psi ,w[i]))
    return psi_bands   

from sympy.utilities.lambdify import lambdify

def eigenfunsH0_Sym_Eval( q, v0, x):
    psi_bands_Sym = eigenfunsH0_Sym( q, v0)
    psi_bands = [] 
    for psi in psi_bands_Sym:
        numpsi = np.array([ complex(sym.N( psi[0], subs = { x_Sym:xval, k_Sym:k, qq_Sym:q } ) )  for xval in x ] )
        psi_bands.append(( numpsi, psi[1] ))
    return psi_bands
```

# Calculation of Wannier states

## The wannier states can be defined with the numerical eigenfunctions


```python
def wannier( x, x0, depth , Nqs=80):
    qs = np.arange(-Nqs+1, Nqs+1) / Nqs
    wfun = np.zeros_like(x,dtype=np.complex) 
    for q in qs:
        psi = eigenfunsH0( q, depth, x)[0][0] 
        wfun += np.exp(-1j*q*(x0-0.5)*np.pi)*psi /  len(qs)
    return wfun    
```

# Calculation of band structure for a 3D lattice 


In a 3D optical lattice the hamiltonian is separable in the three cartesian coordinates.  So we can simply add up the energies and form products of the states from the solution of the 1D problem. 

```python
def bands3d( v0, NBand=0 ):
    # In the case of the 3D band v0 is a vector in 3D
    assert len(v0)==3
    order = np.argsort( v0 )
    l=0; m=0; n=0
    while NBand > 0:
        l+=1; NBand-=1
        if NBand > 0:
            m+=1; NBand-=1
        if NBand > 0:    
            n+=1; NBand-=1
    band1dindex = np.array([ l, m, n] )
        
    bandbot = 0. 
    bandtop = 0. 
    for i,o in enumerate(order):
        in1d = band1dindex[i]
        if in1d%2 ==0:
            bandbot += eigenvalsH0(0., v0[o])[in1d]
            bandtop += eigenvalsH0(1., v0[o])[in1d]
        else:
            bandbot += eigenvalsH0(1., v0[o])[in1d]
            bandtop += eigenvalsH0(0., v0[o])[in1d]
    return np.array([bandbot, bandtop])


def bands3dvec( v0, NBand=0):
    band = numpy.empty( (len(v0), 2) ) 
    for i,v in enumerate(v0):
        band[i] = bands3d(v, NBand)
    return band


def hosc3d( NBand, v0 ):
    return 2* np.sqrt(  v0 ) * (NBand + 1.5)
```

```python
fig = plt.figure(figsize=(6.5,3.8))
gs = matplotlib.gridspec.GridSpec( 1,1)

ax0 = fig.add_subplot( gs[0,0])

for ax in [ax0]:
    ax.set_xlabel("$V_{0} (E_{R})$",fontsize=16)
    ax.grid()
    ax.set_ylim(0.,15.)
    ax.set_xlim(0.,20.)
    
ax0.set_ylabel("$E_{R}$",fontsize=16,rotation=0)

v0 = np.linspace(0.1,20,80.)
v03d = np.outer( v0, np.ones(3))
c = ['blue', 'red', 'green']
fc = ['lightblue','pink','limegreen']
for i in range(3):
    Eb = bands3dvec(v03d,i)
    ax.fill_between( v0 , Eb[:,0], Eb[:,1], \
          color=c[i], facecolor=c[i], alpha=0.7, zorder=4)
    ax.plot( v0, Eb[:,1], color=c[i])
    ax.plot( v0, Eb[:,0], color=c[i],\
          label="Band = %d"%i)
    if i==2: labelstr = 'h.osc.'
    else: labelstr = None
    ax.plot( v0, hosc3d(i,v0), color='black', label=labelstr, zorder=1 )

ax0.legend( bbox_to_anchor=(1.03,1.00), \
            loc='upper left', numpoints=1, \
             prop={'size':12}, handlelength=1.1, handletextpad=0.5 )

gs.tight_layout(fig, rect=[0.,0.,0.8,1.0])
fig.savefig('BandStructure_figures/bands3d_V0.png',dpi=240)
fig.savefig('BandStructure_figures/bands3d_V0.pdf')
```

# Define an interpolation function to quickly calculate the band structure for a 3D lattice 


The interpolation data is saved to file for easy retrieval without having to evaluate the entire thing

```python
v0 = np.linspace(0.,20,80.)
# NPoints = 5
evals0 = np.empty( (len(v0), 2*NPoints+1) )
evals1 = np.empty( (len(v0), 2*NPoints+1) )
for i,v in enumerate(v0):
    evals0[i] = eigenvalsH0(0., v)
    evals1[i] = eigenvalsH0(1., v)

np.savetxt('interpdat_B3D_v0.dat', v0)
for n in range( 2*NPoints+1 ):
    np.savetxt('interpdat_B3D_0_%d.dat'%n, evals0[:,n] )
    np.savetxt('interpdat_B3D_1_%d.dat'%n, evals1[:,n] )
    
```

```python
#Here the interpolation data is loaded from disk
v0 = np.loadtxt('interpdat_B3D_v0.dat')
from scipy.interpolate import interp1d
interp0 = []
interp1 = []
for n in range( 2*NPoints+1 ):
    interp0.append( interp1d(v0, np.loadtxt('interpdat_B3D_0_%d.dat'%n) ))
    interp1.append( interp1d(v0, np.loadtxt('interpdat_B3D_1_%d.dat'%n) ))
    
def bands3dinterp( v0, NBand=0 ):
    # In the case of the 3D band v0 is a vector in 3D
    assert len(v0)==3
    order = np.argsort( v0 )
    l=0; m=0; n=0
    while NBand > 0:
        l+=1; NBand-=1
        if NBand > 0:
            m+=1; NBand-=1
        if NBand > 0:    
            n+=1; NBand-=1
    band1dindex = np.array([ l, m, n] )
        
    bandbot = 0. 
    bandtop = 0. 
    for i,o in enumerate(order):
        in1d = band1dindex[i]
        if in1d%2 ==0:
            bandbot += interp0[in1d](v0[o]) 
            bandtop += interp1[in1d](v0[o])
        else:
            bandbot += interp1[in1d](v0[o])
            bandtop += interp0[in1d](v0[o])
    return np.array([bandbot, bandtop])


def bands3dinterpvec( v0, NBand=0):
    band = numpy.empty( (len(v0), 2) ) 
    for i,v in enumerate(v0):
        band[i] = bands3dinterp(v, NBand)
    return band
    
```

# Calculation of tunneling coefficient


It is useful to compare a tight-binding dispersion relation with the exact form.  They only differ significantly for very shallow lattices.

```python
v0 = 2.0
qs = np.linspace(-1,1,40)
plt.plot( qs, bandH0( 0, qs, v0, NPoints=5))
E0 = bandH0(0, [0.], v0)[0]
E1 = bandH0(0, [1.], v0)[0]
#plt.plot( qs, E0*np.ones_like(qs) )
#plt.plot( qs, E1*np.ones_like(qs) )
plt.plot( qs , E1 - (E1-E0)*np.cos(qs*np.pi)/2 - (E1-E0)/2)
qs2 = np.linspace(-2,2,80)
#plt.plot( qs2, np.cos(qs2*np.pi)**2 )
```

```python
def tCalc(v0, Xij=1., Nqs=80):
    qs = np.arange(-Nqs+1, Nqs+1) / Nqs
    t = 0.
    for q in qs:
        t += eigenvalsH0(q, v0)[0] * np.exp( 1j*np.pi*q*Xij)  / len(qs)
    return -1*np.real(t)
```
