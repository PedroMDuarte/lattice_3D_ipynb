```python
from compensated import *

matplotlib.rc('text', usetex=True)
#matplotlib.rc('font',**{'family':'serif'})

import os

basedir = 'Inhomog'

def fname(IR,GR,Na,Ti,sub,aS,ext):
    fpath = (basedir+'/{0:.2f}_{1:.2f}_{2:.1e}/T{3:.1f}/'.format(IR,GR,Na,Ti) + sub + '_%04da0'%aS).replace('.','p') + ext
    dpath = os.path.dirname(fpath)
    if not os.path.exists(dpath):
        os.makedirs(dpath)
    return fpath

def tname(IR,GR,Na,Ti):
    return os.path.split(os.path.dirname( fname(IR,GR,Na,Ti,'',0.,'')))[0] + '/T%02d_PhaseDiagram.dat'%Ti

def pname(IR,GR,Na):
    return os.path.split(os.path.dirname( fname(IR,GR,Na,0,'',0.,'')))[0]

#print fname(7.0, 3.85, 3.5e5, 2,'COL',80.,'.png')
#print tname(7.0, 3.85, 3.5e5, 2)
#print pname(7.0, 3.85, 3.5e5)

import glob
import re
def geti(datf):
    Ti =  re.search(r'T[0-9]{2}_',datf).group()
    return int(re.search(r'[0-9]{2}',Ti).group())
```

```python
IR = 5.5
GR = 2.5
Na = 3.3e5
#aS = 326.
Tt = 1.8

# Calibrated beam waists
waists = ((49.22,49.22), (49.95,49.95), (44.34,44.34))
GRwaists = ((43.8,43.8), (40.8,40.8), (39.2,39.2))



# U/t values for 5.5 Er data and their corresponding scattering lengths
aS = { 1.636:80. , 4.976:244. , 7.771:380. , 9.713:475. , 11.860:580. , 13.905:680. , 16.359:801 } 

peakd = [] ; rad = [] ; peakt = []
dens = [] ; Ut = [] ; U=[];  t=[];  D = []
for u in sorted(aS.keys()):
#for u in [11.860]:
    a = aS[u]
    latt3d = simpleCubic( allIR=IR, allGR=GR, Natoms=Na, a_s=a, Temperature=Tt,\
             waists = waists, green_waists = GRwaists )

    #fig111 = latt3d.plot3Line(func=latt3d.Bands,  lines='111');
    fig111, dens_binned, Ut_binned, U_binned, t_binned, peak_dens, radius1e, peak_t = latt3d.CheckInhomog( closefig = True ) ; 
    
    dens.append( dens_binned ) 
    Ut.append( Ut_binned ) 
    U.append( U_binned )
    t.append( t_binned )
    
    peakd.append(peak_dens) 
    peakt.append(peak_t)
    rad.append( radius1e ) 
    D.append( latt3d.getNumberD()/latt3d.Number )

    fig111.savefig(fname(IR,GR,Na,Tt,'111',a,'.png'), dpi=200)
    
scattlen = np.array( [aS[a] for a in sorted(aS.keys()) ] )
np.savetxt('Inhomog/SampleParams/gr_%.1f_.dat'%GR, np.column_stack(( scattlen, np.array(peakd), np.array(D), np.array(rad) )), fmt='%.3g' ) 
```

```python
from matplotlib import rc
rc('text', usetex=True)


fig = plt.figure(figsize=(10.,3.5))
gs = matplotlib.gridspec.GridSpec( 1,3)
axL = [ fig.add_subplot( gs[0,i] ) for i in range(3) ] 

import glob
for fdat in sorted(glob.glob('Inhomog/SampleParams/*dat')):
    dat = np.loadtxt(fdat)
    for i in range(3):
        axL[i].plot( dat[:,0], dat[:,1+i], '-', marker='.', label=fdat.split('_')[1]+' $E_{R}$') 
        for l in axL[i].xaxis.get_ticklabels():
            l.set_rotation(30)
        axL[i].grid()

    
axL[1].set_xlabel('Scattering length') 
axL[0].set_ylabel('Peak density (atoms per site)')
axL[1].set_ylabel('Double occupancy')
axL[2].set_ylabel('$1/e$ size')
    
gs.tight_layout(fig, rect=[0.,0.,0.9,1.0])
axL[2].legend( bbox_to_anchor=(1.03,1.00), \
        loc='upper left', numpoints=1, \
         prop={'size':12}, handlelength=1.1, handletextpad=0.5 )

fig.savefig('Inhomog/SampleParams/GR.png',dpi=240)
```

```python
Tstar = []
Ustar = []
for i,a in enumerate(sorted(aS.keys())):
    peakU = a * peakt[i]
    print "-------------------------------"
    print "At center:"
    print "  t = %.3f Er\n  U = %.3f Er\nU/t = %.3f Er" %(peakt[i], peakU , peakU/peakt[i])
    print "\nAt bins:"
    print "bin#    nb   t/tb   Ub/tb     u     "
    print "-------------------------------"
    Tstar_ = []; Ustar_ = []; 
    for j in range(len(dens[i])):
        Tstar_.append( peakt[i]/t[i][j] ) 
        u = Ut[i][j] / a
        Ustar_.append( u ) 
        print "%03d%7.2f%7.2f%7.2f%7.2f"% ( j, dens[i][j], peakt[i]/t[i][j], U[i][j]/t[i][j], u)
    print "\n"
    Tstar.append( Tstar_ )
    Ustar.append( Ustar_ ) 
    
```
