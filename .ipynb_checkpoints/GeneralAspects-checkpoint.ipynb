```python
from matplotlib import rc
rc('font',**{'family':'serif'})
rc('text', usetex=True)

fig = plt.figure(figsize=(6.5,2.9))
gs = matplotlib.gridspec.GridSpec( 1,1, \
        wspace=0.25, left=0.1, right=0.48, bottom=0.18, top=0.94)
ax = fig.add_subplot( gs[0,0])

def vL( s0, wL, r ):
    return s0*np.exp( - 4*r**2 / ( 3. * wL**2 ) ) 

def vLenv( s0, wL, r ):
    return -3.*vL( s0, wL, r) 

def vLenvTop( s0, wL, r):
    return vLenv( s0, wL, r) + vL( s0, wL, r)

def vLatt( s0, wL, r):
    return vLenv( s0, wL, r) + vL( s0, wL, r) *  ( sin(2.*np.pi * r111 / 1.064 /  10. )**2 )

def band_bot( s0, wL, r):
    return vLenv( s0, wL, r) + 3.*np.sqrt( vL( s0, wL, r) ) 

def band_top( s0, wL, r):
    return vLenv( s0, wL, r) + 3.*np.sqrt( vL( s0, wL, r) )  \
           + 48./np.sqrt(np.pi) * ( vL(s0,wL,r)**(3./4.)) * np.exp( -2.*np.sqrt( vL(s0,wL,r)) )
    
def band_1st( s0, wL, r):
    return vLenv( s0, wL, r) + 5.*np.sqrt( vL( s0, wL, r) )  

    
r111 = np.linspace( -80., 80., 540 ) 

s0 = 10.
wL = 47. 

ax.plot( r111, vLenv( s0, wL, r111 ), color='black', lw=2. , \
        label='$V_{L,\mathrm{env}}(r_{\mathrm{d}})$') 
ax.plot( r111, vLenvTop( s0, wL, r111 ), '--',color='black', lw=2., \
        label='$V_{L,\mathrm{env}}(r_{\mathrm{d}}) + s(r_{\mathrm{d}})$' ) 
ax.plot( r111, vLatt( s0, wL, r111 ), color='gray', lw=0.8, alpha=0.5, label=r'$\mathrm{Lattice}\ \lambda\!\times\!\!10$' ) 
ax.plot( r111, band_bot( s0, wL, r111), color='blue', lw=1.4, label=r'$E_{0}  = V_{L,\mathrm{env}}(r_{\mathrm{d}}) +\frac{3}{2} \hbar \omega_{0} $')
ax.plot( r111, band_top( s0, wL, r111), color='blue', lw=1.4, label=r'$E_{0} + W = V_{L,\mathrm{env}}(r_{\mathrm{d}}) +\frac{3}{2} \hbar \omega_{0} + W$')

ax.fill_between( r111, band_bot( s0, wL, r111), band_top( s0, wL, r111), color='blue', alpha=0.5) 

ax.plot( r111, band_1st( s0, wL, r111), color='red', lw=1.4, label=r'$E_{1} = V_{L,\mathrm{env}}(r_{\mathrm{d}}) +\frac{5}{2} \hbar \omega_{0} $')
ax.axhline( band_1st( s0, wL, 0.), color='red', lw=0.8, alpha=0.3)
#ax.axvline( x=24.2, ymin=band_1st( s0, wL, 0.), ymax=-5., color='red', lw=0.8, alpha=0.5) 
ax.axvline( x=22.2,  color='red', lw=0.8, alpha=0.3) 

ax.annotate("",
            xy=(24.2, -5.), xycoords='data',
            xytext=(-1.5, -5.), textcoords='data',
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="arc3",color='red'), color='red'
            )
ax.text( 10., -4., '$r_{1}$', ha='center', va='center', color='red')

ax.grid()


ax.set_ylabel('$E_{R}$')
ax.set_xlabel('$\mu\mathrm{m\ along\ 111}$')
ax.set_xlim(-80.,80.)

ax.legend( bbox_to_anchor=(1.03,1.00), \
        loc='upper left', numpoints=2, \
        prop={'size':12}, handlelength=2.2, handletextpad=0.5 )

fig.savefig('GeneralAspects_figures/lattice_general.png', dpi=300)
```

## Maximum locked number vs.  lock depth.  For various lattice beam waists 

```python
from matplotlib import rc
rc('font',**{'family':'serif'})
rc('text', usetex=True)

fig = plt.figure(figsize=(3.8,2.9))
gs = matplotlib.gridspec.GridSpec( 1,1, \
        wspace=0.25, left=0.25, right=0.95, bottom=0.15, top=0.95)
ax = fig.add_subplot( gs[0,0])


from scipy import optimize

def band_diff( s0, wL, r111):
    return band_bot( s0, wL, r111) - band_bot( s0, wL, 0.) \
        - ( band_1st( s0, wL, 0. )- band_bot( s0, wL, 0.) )

def find_r1( s0, wL ):
    f = lambda x : band_diff( s0, wL, x ) 
    return optimize.brentq( f , 0., 200., )


def band_diff_comp( s0, wL, g0, wC,r111):
    return band_bot( s0, wL, r111) - band_bot( s0, wL, 0.) \
              + 3.*vL( g0, wC, r111) - 3.*vL( g0, wC, 0.) \
        - ( band_1st( s0, wL, 0. )- band_bot( s0, wL, 0.) )
def find_r1_comp( s0, wL, g0, wC):
    f = lambda x : band_diff_comp( s0, wL, g0, wC, x ) 
    return optimize.brentq( f , 0., 400., )


ss = np.linspace( 10., 60., 20 ) 

wLs = [47., 54., 62.,70.]
for wL in wLs:
    r1 = []
    for s in ss:
        r1.append( find_r1( s,wL))
    r1 = np.array( r1 ) 
    ax.plot(  ss,  4./3. * np.pi * ( r1/ (1.064/2.) )**3, label='$w_{L}=%d\,\mu\mathrm{m}$'%wL, lw=2 )
    
wC = 40. 
wLs = [47.]
for wL in wLs:
    r1 = []
    for s in ss:
        r1.append( find_r1_comp( s,wL, 3.0, wC))
    r1 = np.array( r1 ) 
    ax.plot(  ss,  4./3. * np.pi * ( r1/ (1.064/2.) )**3, ':', label='$w_{L}=%d\,\mu\mathrm{m},\ \mathrm{comp.}$'%wL, lw=2 )
    
    
ax.grid()
ax.set_ylim( 0., 500000)

ax.set_xlabel('$\mathrm{Lattice\ depth\ } (E_{R})$')  
ax.set_ylabel(r'$N_{1} =  \frac{4}{3} \pi \left( \frac{ r_{1} }{ a  }  \right)^{3}$', fontsize=12)

ax.legend( bbox_to_anchor=(1.05,1.00), \
        loc='upper right', numpoints=1, \
        prop={'size':8}, handlelength=1.1, handletextpad=0.5 )


fig.savefig('GeneralAspects_figures/lattice_lock.png', dpi=300)
```

## Characteristic energy vs.  beam waist for 500,000 atoms

```python
from matplotlib import rc
rc('font',**{'family':'serif'})
rc('text', usetex=True)

fig = plt.figure(figsize=(3.2,2.6))
gs = matplotlib.gridspec.GridSpec( 1,1, \
        wspace=0.25, left=0.18, right=0.95, bottom=0.18, top=0.9)
ax = fig.add_subplot( gs[0,0])

def alpha( s0, wL):
    return ( 4*s0 - 2*np.sqrt(s0) ) / ( (wL/(1.064/2.))**2.)  

def t( s0 ):
    return 4./np.sqrt(np.pi) * ( s0**(3./4.)) * np.exp( -2.*np.sqrt( s0) )

def Et( s0, wL, Natoms):
    return alpha(s0,wL) * (  3.*Natoms / (4.*np.pi) )**(2./3.)  / t(s0 ) 


wLs = np.linspace( 60., 200., 100 )
#plt.plot( wLs,  alpha( 10., wLs) ) 
ax.plot( wLs, Et( 10., wLs, 500000 ), lw=3) 

ax.set_xlabel('$w_{L}\ (\mu\mathrm{m})$')
ax.set_ylabel('$E_{\mathrm{t}} / t $')
ax.grid()
ax.axhline( 40., color='black', alpha=0.6)

fig.savefig('GeneralAspects_figures/Et-waist.png', dpi=190)
```

## Compensated lattice analytical profiles

```python
from matplotlib import rc
rc('font',**{'family':'serif'})
rc('text', usetex=True)

fig = plt.figure(figsize=(5.5,2.9))
gs = matplotlib.gridspec.GridSpec( 1,1, \
        wspace=0.25, left=0.12, right=0.55, bottom=0.18, top=0.9)
ax = fig.add_subplot( gs[0,0])

def vL( s0, wL, r ):
    return s0*np.exp( - 4*r**2 / ( 3. * wL**2 ) ) 

def vLenv( s0, wL, r ):
    wC = 40. 
    g0 = (s0 - np.sqrt(s0)/2.)/(47./40.)**2
    #print "g0 = ",g0
    
    return -3.*vL( s0, wL, r)  + 3.*g0*np.exp( - 4*r**2 / ( 3. * wC**2 ) ) 

def vLenvTop( s0, wL, r):
    return vLenv( s0, wL, r) + vL( s0, wL, r)

def vLatt( s0, wL, r):
    return vLenv( s0, wL, r) + vL( s0, wL, r) *  ( sin(2.*np.pi * r111 / 1.064 /  10. )**2 )

def band_bot( s0, wL, r):
    return vLenv( s0, wL, r) + 3.*np.sqrt( vL( s0, wL, r) ) 

def band_top( s0, wL, r):
    return vLenv( s0, wL, r) + 3.*np.sqrt( vL( s0, wL, r) )  \
           + 48./np.sqrt(np.pi) * ( vL(s0,wL,r)**(3./4.)) * np.exp( -2.*np.sqrt( vL(s0,wL,r)) )
    
def band_1st( s0, wL, r):
    return vLenv( s0, wL, r) + 5.*np.sqrt( vL( s0, wL, r) )  

def quartic( s0, wL, r):
    alpha = 47./40.
    zero = 3.*( np.sqrt(s0)*(1. - 1./(2.*alpha**2)) + s0*( 1./alpha**2 - 1. ))
    return zero + 2.*(np.sqrt(s0)*(1.-2.*alpha**2) + 4.*s0*(1.-alpha**2)) / ( 3. *wL**4 )  * r**4
    
    

    
r111 = np.linspace( -80., 80., 540 ) 

s0 = 10.
wL = 47. 

ax.plot( r111, vLenv( s0, wL, r111 ), color='black', lw=2. , \
        label='$V_{\mathrm{env}}(r_{\mathrm{d}})$') 
ax.plot( r111, vLenvTop( s0, wL, r111 ), '--',color='black', lw=2., \
        label='$V_{\mathrm{env}}(r_{\mathrm{d}}) + s(r_{\mathrm{d}})$' ) 
ax.plot( r111, vLatt( s0, wL, r111 ), color='gray', lw=0.8, alpha=0.5, label=r'$\mathrm{Lattice}\ \lambda\!\times\!\!10$' ) 
ax.plot( r111, band_bot( s0, wL, r111), color='blue', lw=1.4, label=r'$V_{\mathrm{env}}(r_{\mathrm{d}}) +\frac{3}{2} \hbar \omega_{0} $')
ax.plot( r111, band_top( s0, wL, r111), color='blue', lw=1.4, label=r'$V_{\mathrm{env}}(r_{\mathrm{d}}) +\frac{3}{2} \hbar \omega_{0} + W$')

ax.fill_between( r111, band_bot( s0, wL, r111), band_top( s0, wL, r111), color='blue', alpha=0.5) 

ax.plot( r111, band_1st( s0, wL, r111), color='red', lw=1.4, label=r'$V_{\mathrm{env}}(r_{\mathrm{d}}) +\frac{5}{2} \hbar \omega_{0} $')

ax.plot( r111, quartic( s0,wL,r111), color='green', lw=2.)

ax.grid()

ax.set_ylim(-20.,10.)


ax.set_ylabel('$E_{R}$')
ax.set_xlabel('$\mu\mathrm{m\ along\ 111}$')

ax.legend( bbox_to_anchor=(1.01,1.00), \
        loc='upper left', numpoints=2, \
        prop={'size':12}, handlelength=2.2, handletextpad=0.5 )

fig.savefig('GeneralAspects_figures/lattice_general-comp.png', dpi=300)
```

## Make a plot that shows the optimal waist ratio for evaporation as a function of lattice depth. 

```python
from matplotlib import rc
rc('font',**{'family':'serif'})
rc('text', usetex=True)

fig = plt.figure(figsize=(4.2,2.6))
gs = matplotlib.gridspec.GridSpec( 1,1, \
        wspace=0.25, left=0.14, right=0.65, bottom=0.18, top=0.9)
ax = fig.add_subplot( gs[0,0])

def tunnel( s ):
    return 4./np.sqrt(np.pi) * ( s**(3./4.)) * np.exp( -2.*np.sqrt( s) ) 

def optimal_alpha( s0, U):
    # U is given in units of t, so here we convert it to recoil
    Urec = U * tunnel(s0) 
    return  np.sqrt( (4.*s0 - 2.*np.sqrt(s0) ) / ( 4.*s0 - 4.*np.sqrt(s0) - Urec )  )

s0s = np.linspace( 5., 10. , 100 ) 
Us = [0., 10, 20., 30. ]
for U in Us:
    ax.plot( s0s, optimal_alpha( s0s, U ) , lw=2, label='$U/t=%d$'%U  )
            
ax.set_xlabel(r'$s_{0}$')
ax.set_ylabel(r'$\alpha_{\mathrm{evap}}$', fontsize=14)
ax.grid()


ax.legend( bbox_to_anchor=(1.01,1.00), \
        loc='upper left', numpoints=2, \
        prop={'size':12}, handlelength=2.2, handletextpad=0.5 )

fig.savefig('GeneralAspects_figures/alpha-evap-optimal.png', dpi=300)
```

## Comparison of the quartic potential with its series expansions

```python
from matplotlib import rc
rc('font',**{'family':'serif'})
rc('text', usetex=True)

fig = plt.figure(figsize=(5.5,2.9))
gs = matplotlib.gridspec.GridSpec( 1,1, \
        wspace=0.25, left=0.12, right=0.55, bottom=0.18, top=0.9)
ax = fig.add_subplot( gs[0,0])

def vL( s0, wL, r ):
    return s0*np.exp( - 4*r**2 / ( 3. * wL**2 ) ) 

def vC( g0, wC, r ):
    return g0*np.exp( - 4*r**2 / ( 3. * wC**2 ) ) 

def band( s0, g0, wL, wC, r):
    return -3.*vL( s0, wL, r) + 3.*vC( g0, wC, r) + 3.*np.sqrt( vL(s0, wL, r ) )



alpha = 1.17
s0 = 7.
wL = 47. 
wC = wL/alpha
g0 = (4.*s0 - 2.*np.sqrt(s0))/(4.*alpha**2) 

print "g0 = %.2f"%g0


r111 = np.linspace( -80., 80., 540 ) 

# Plot the actual function
ax.plot( r111, band(s0, g0, wL, wC, r111 ), lw=2. ,\
        label='$E_{0}$')
ax.axhline( g0 + np.sqrt(s0) - s0, lw=1, color='black', label='$E_{\mathrm{th}}$')

# Define and plot the series exansions
import sympy.mpmath as mpmath

# Need to redefine the functions with mpmath functions
def mp_vL( s0, wL, r ):
    return s0*mpmath.exp( - 4*r**2 / ( 3. * wL**2 ) ) 
def mp_vC( g0, wC, r ):
    return g0*mpmath.exp( - 4*r**2 / ( 3. * wC**2 ) ) 
def mp_band( s0, g0, wL, wC, r):
    return -3.*mp_vL( s0, wL, r) + 3.*mp_vC( g0, wC, r) + 3.*mpmath.sqrt( mp_vL(s0, wL, r ) )

for i in [4, 6, 8]:
    def f(x):
        return mp_band(s0,g0,wL,wC,x)
    p = mpmath.taylor( f, 0., i ) 
    def eval_series(r):
        return mpmath.polyval(p[::-1], r)
    
    ax.plot( r111, eval_series(r111), '--', lw=2, label='%d'%i)



ax.set_ylim(-1., 2.)
ax.grid()
ax.set_ylabel('$E_{R}$')
ax.set_xlabel('$\mu\mathrm{m\ along\ 111}$')

ax.legend( bbox_to_anchor=(1.01,1.00), \
        loc='upper left', numpoints=2, \
        prop={'size':12}, handlelength=2.2, handletextpad=0.5 )

#fig.savefig('GeneralAspects_figures/lattice_general-comp.png', dpi=300)
```
