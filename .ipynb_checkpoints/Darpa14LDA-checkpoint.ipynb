## Comparison between traditional and compensated lattice setup

```python
import compensated
latt3d = compensated.simpleCubic(\
         waists = ((150.,150.),(150.,150.),(150.,150.)),
         allIR=5.5, allGR=0., Natoms=3.52e5, a_s=650.)
fig111 = latt3d.PlotBandsSimple( )
latt3d.get_eta_evap()
fig111.savefig('/home/pmd/sandbox/darpa14/LDA/traditional.png', dpi=300 )

fig111 = latt3d.plot3Line( func=latt3d.Bands, lines='111')
fig111.savefig('/home/pmd/sandbox/darpa14/LDA/traditional_111.png', dpi=300 )

fig111 = latt3d.plot3Line( func=latt3d.Bands, lines='111', y2lims=(0.,10.), entropy_per_particle=True)
fig111.savefig('/home/pmd/sandbox/darpa14/LDA/traditional_111_entropy.png', dpi=300 )
```

```python
import compensated
latt3d = compensated.simpleCubic( allIR=5.5, allGR=2.6, Natoms=3.515e5, a_s=650.)
fig111 = latt3d.PlotBandsSimple( )
latt3d.get_eta_evap()
fig111.savefig('/home/pmd/sandbox/darpa14/LDA/compensated.png', dpi=300 )

fig111 = latt3d.plot3Line( func=latt3d.Bands, lines='111')
fig111.savefig('/home/pmd/sandbox/darpa14/LDA/compensated_111.png', dpi=300 )

fig111 = latt3d.plot3Line( func=latt3d.Bands, lines='111', y2lims=(0.,10.), entropy_per_particle=True)
fig111.savefig('/home/pmd/sandbox/darpa14/LDA/compensated_111_entropy.png', dpi=300 )
```

## Make plots of Doublon fraction and density as a function of U/t for varios  T/t in the LDA

```python
scattlen = np.linspace( 80., 1200., 28)
Ts = [1.6, 2.0, 2.4, 3.2 ] 

for j,T in enumerate(Ts):
    print "\nWorking on T = %.1f ..." % T
    docc = []
    peakd = []
    for i,a in enumerate(scattlen):
        print a, " ",
        
        #figCol = latt3d.column( ["density","doublons","entropy"] )
        #figCol.savefig('Ut_Comp/T%02d/Col_as%04d.png'%(Ti,a), dpi=200)
        #fig111 = latt3d.plot3Line(func=latt3d.Bands,  lines='111')
        #fig111.savefig('Ut_Comp/T%02d/111_as%04d.png'%(Ti,a), dpi=200)
        
        latt3d = compensated.simpleCubic( allIR=5.5, allGR=2.75, Natoms=3.5e5, a_s=a, Temprature=T)
        docc.append( latt3d.getNumberD()/latt3d.Number )
        peakd.append( np.amax(latt3d.fHdens( latt3d.onsite_t_111, latt3d.localMu_t_111 ))) 
        
        
    fname = '/home/pmd/sandbox/darpa14/LDA/dat/T%02d.dat' %  (T*10.)
    print "Saving to ", fname
    np.savetxt(fname , \
            np.column_stack(( scattlen, np.array(docc), np.array(peakd) ) ))
```
