## Comparison for N=350,000 atoms 

```python
T = 1.8

waist0 = 135.
latt3d_00 = compensated.simpleCubic(\
         waists = ((waist0,waist0),(waist0,waist0),(waist0,waist0)),
         allIR=5.5, allGR=0., Natoms=3.52e5, a_s=650., Temperature=T)

latt3d_01 = compensated.simpleCubic( allIR=5.5, allGR=2.6, Natoms=3.515e5, a_s=650., Temperature=T)

latt3d_02 = compensated.simpleCubic( allIR=5.5, allGR=0., Natoms=3.50e5, a_s=650., Temperature=T)

latt3d_03 = compensated.simpleCubic( allIR=5.5, allGR=0., Natoms=.17e5, a_s=650., Temperature=T)

latts = [ latt3d_00, latt3d_01, latt3d_02, latt3d_03 ] 
```

```python
fig = plt.figure(figsize=(8.,8.))
gs = matplotlib.gridspec.GridSpec(2,2)
axs =  [ 
        fig.add_subplot( gs[0,0] ),\
        fig.add_subplot( gs[0,1] ),\
        fig.add_subplot( gs[1,0] ),\
        fig.add_subplot( gs[1,1] ),\
        ]


qtys = [
        'density',\
        'doublons',\
        'entropy',\
        'entropy_per_particle'\
        ]


for l,latt in enumerate(latts):
    #print 
    #latt.get_eta_evap()
    for q,qty in enumerate(qtys):
        latt.plotLineQty( ax=axs[q], qty=qty, plotkwargs=dict(lw=2,label='%d'%l))

for ax in axs:
    ax.set_xlim( -50., 50. )
    ax.legend( bbox_to_anchor = (1.05,1.05), loc='upper right', \
            numpoints=1, labelspacing=0.2, prop={'size':10},\
            handlelength=1.1, handletextpad=0.5 )
    
axs[3].set_ylim(0., 5.);
```

### Observations 

In the plot above one can see that the smaller beam waist, which causes an inhomogeneity in the Hubbard parameters does not provide an advantage when it comes to entropy redistribution.  

In fact the entropy per particle is most sensitive to density rather than to t, U, U/t  so it is the density inhomogeneity that does most of the entropy redistribution.  

For a given number of atoms one just has to find the confinement that gives 1 per site.   In a large beam waist lattice this is done by adding additional confinement beams to squeeze the sample down.   In our setup, this is done by adding anti-confinement beams


We also plotted above a lattice with small beam waists and no compensation, but with a low enough number of atoms such that n=1 at the center.  The number of atoms that is needed is very low, thus it produces a very small Mott plateau.   This is what Mathy, Huse, Hulet refer to as enlarging the Neel state.   

In my opinion this kind of sucks because it's all about the atom number that is required to get n=1 at the center. 

### Comparisons for different beam waists

In the comparison above we used the same number of atoms.  For the uncompensated sample we found the lattice waists such that the filling at the center would be n=1.   For the uncompensated sample we used our IR beam waist and found the compensation such that the filling at the center would be n=1.   

A filling of n=1 can be satisfied pretty much all across the Mott gap,  so it would be nice to define a better criteria.  For example having a global chemical potential that is right in the center of the Mott gap.  

Below we use the keyword argument globalMu = 'halfMott'  in order to set the chemical potential at the center of the Mott gap.   This is going to allow us to make quick assesments of different kinds of setups

```python
import compensated
T = 1.8

waist0 = 134.
latt3d_00 = compensated.simpleCubic(\
         waists = ((waist0,waist0),(waist0,waist0),(waist0,waist0)),
         allIR=5.5, allGR=0., globalMu='halfMott', a_s=650., Temperature=T)
print latt3d_00.EtaEvap
fig111 = latt3d_00.PlotBandsSimple()

latt3d_01 = compensated.simpleCubic( allIR=5.5, allGR=2.62, globalMu='halfMott', a_s=650., Temperature=T)
print latt3d_01.EtaEvap
fig111 = latt3d_01.PlotBandsSimple( )
```

### Now let's pick a beam waist and vary the compensation to see what happens

```python
import compensated
T = 1.8

def vary_comp( comps, wIR=47., wGR=40, s=5.5 ):
    fig = plt.figure(figsize=(8.,4.))
    gs = matplotlib.gridspec.GridSpec(1,2)
    axs =  [ 
            fig.add_subplot( gs[0,0] ),\
            fig.add_subplot( gs[0,1] ),\
            ]
    
    
    etas = []
    nums = [] 
    for comp in comps:
        latt3d_01 = compensated.simpleCubic( allIR=s, allGR=comp, \
                        waists = ((wIR,wIR),(wIR,wIR),(wIR,wIR)),
                        green_waists = ((wGR,wGR),(wGR,wGR),(wGR,wGR)),
                        globalMu='halfMott', a_s=650., Temperature=T)
        
        if latt3d_01.bandBot_LocalMin is False:
            print "Warning GR=%.2f makes a hump at the center"%comp
        etas.append( latt3d_01.EtaEvap ) 
        nums.append( latt3d_01.Number ) 
        
    etas = np.array(etas)
    nums = np.array(nums)
    
    print "N=%e,  eta=%.2f"%( nums[-1], etas[-1] ) 
        
    axs[0].plot( comps, etas, marker='o')
    axs[1].plot( comps, nums, marker='o')

```
