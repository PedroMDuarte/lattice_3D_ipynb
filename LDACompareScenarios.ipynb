### We define a function to automate the procedure of finding the value of the compensation such that  n=1 at the center and N=350,000

```python
from scipy import optimize
import compensated
T = 1.8

def EtaEvap(  s, wIR, wGR, save= False, savefile='LDACompare_figures/default.png', AtomNum=350000.):
    """ This function finds the eta parameter for evaporation that is obtained
        at half-filling and for 350,000 atoms.  
        
        For a certain value of the IR and Green beam waists you cand find a 
        compensation value that will make an n=1 sample with 350,000 atoms.  
        
        At this compenation value we evaluate ETA. """ 
    def latt_number( comp, s, wIR, wGR ):
        try:
            latt3d_01 = compensated.simpleCubic( allIR=s, allGR=comp, \
                                waists = ((wIR,wIR),(wIR,wIR),(wIR,wIR)),
                                green_waists = ((wGR,wGR),(wGR,wGR),(wGR,wGR)),
                                globalMu='halfMott', a_s=650., Temperature=T)
            return latt3d_01.Number
        except:
            return 1e6
    
    fN = lambda x : latt_number( x, s, wIR, wGR ) - AtomNum
    verbose = False
    if verbose:
        print "Searching for compensation such that N=3.5e5..."
    compN, brentResults = optimize.brentq( fN, 0., 6.0, xtol=1e-2, rtol=1e3, full_output=True)
    if verbose:
        print "compN = %.3f, " % brentResults.root,
        print "n_iter=%d, " % brentResults.iterations,
        print "n_eval=%d, " % brentResults.function_calls,
        print "converge=? ", brentResults.converged
        
    latt3d = compensated.simpleCubic( allIR=s, allGR=compN, \
                            waists = ((wIR,wIR),(wIR,wIR),(wIR,wIR)),
                            green_waists = ((wGR,wGR),(wGR,wGR),(wGR,wGR)),
                            globalMu='halfMott', a_s=650., Temperature=T)
    
    if save:
        fig111 = latt3d.PlotBandsSimple( )
        fig111.savefig(savefile)
    return [compN, latt3d.Number, latt3d.EtaEvap, latt3d.Entropy/latt3d.Number]
```
